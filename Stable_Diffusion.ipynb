```python
# トークン変数の準備
YOUR_TOKEN="ここにトークンを入力する"
from diffusers import StableDiffusionPipeline

# StableDiffusionパイプラインの準備
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", use_auth_token=YOUR_TOKEN)
pipe.to("cuda")
```

```python
import torch
prompt = "A concept idea of a KAWAII girl with platinum-straight bangs and a frilly dress on a white background was featured in the pivix fanbox digital art,Full body standing picture.thick coating painting, Bold line painting,Fantastic, symmetrical, mabinogi, Arknights, Shadowverse, Azur Lane, WLOP, Makoto Shinkai, Yoh Yoshinari"
i=0
while i<1:
  image = pipe(prompt,width=512,height=512,num_inference_steps=150)["sample"][0]
  image.save("out"+str(i)+".png")
  i=i+1
```
